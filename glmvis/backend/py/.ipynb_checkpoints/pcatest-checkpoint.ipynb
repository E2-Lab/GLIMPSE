```python
import pandas as pd
import numpy as np
import sys
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
nemb = pd.read_csv("../csv/metrics_123.csv", header=None)
nemb
```

```python
lbl = nemb[0].tolist()
```

```python
principalComponents = pca.fit_transform(nemb)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

print(principalDf)

fig = plt.figure(figsize = (16,16))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title("2 Component PCA", fontsize = 20)
ax.scatter(principalDf['principal component 1'],principalDf['principal component 2'],s=100)

for i, txt in enumerate(lbl):
 ax.annotate(txt, (principalDf['principal component 1'][i]+.02, principalDf['principal component 2'][i]), fontsize=15)
ax.grid()
fig.savefig('../figs/plot.png')
```
